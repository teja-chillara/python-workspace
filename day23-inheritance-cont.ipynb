# single level inheritance

```python
class Father: # super class
  def father_property(self): # instance method
    print("father property method")

class Son: # sub class
  def son_property(self): # instance method
    print("son property method")
```

```python
son = Son()
```

```python
son.son_property()
```

```python
class Father: # super class
  # static variables
  a = 10
  b = 20
  def father_property(self): # instance method
    print("father property method")

class Son(Father): # sub class
  def son_property(self): # instance method
    print("son property method")
```

```python
Son.mro() # object: super most class
```

```python
son = Son()
```

# multi level inheritance

```python
# if you are have inheritance relationship between above 2 classes is called multi level inheritance
```

```python
class GrandFather:
  g = 1000
  def grand_property(self):
    print("grand father property")

class Father(GrandFather):
  f = 10000
  def father_property(self):
    print("father property method")

class Son(Father):
  s = 100000
  def son_property(self):
    print('son property method')
```

```python
g = GrandFather()
```

```python
g.g
```

```python
g.grand_property()
```

```python
f = Father()
```

```python
f.grand_property()
```

```python
f.father_property()
```

```python
s = Son()
```

```python
s.son_property()
```

```python
s.father_property()
```

```python
s.grand_property()
```

# Hierarchial inheritance

image.png



```python
class GrandFather:
  g = 1000
  def grand_property(self):
    print("grand father property")

class Father(GrandFather):
  f = 10000
  def father_property(self):
    print("father property method")

class Son(Father): # brother
  s = 100000
  def son_property(self):
    print('son property method')

class Daughter(Father): # sister
  d = 50000
  def daughter_property(self):
    print('daughter property')
```

```python
s = Son()
```

son class can access only:
1. his own properties
2. fathers
3. grand fathers

```python
d = Daughter()
```

```python
d.daughter_property()
```

```python
d.father_property()
```

```python
d.grand_property()
```

# mutltiple inheritance

image.png

```python
class Father:
  f = 50000
  def father_method(self):
    print('father class method')

class Mother:
  m = 70000
  def mother_method(self):
    print("mother class method")

class Son(Father, Mother): # multiple inheritance
  s = 298349
  def son_method(self):
    print('son class method')
```

```python
s = Son()
```

```python
s.son_method()
```

```python
s.father_method()
```

```python
s.mother_method()
```

```python
class Father:
  f = 50000
  def father_method(self):
    print('father class method')

  def joint_property(self):
    print("father share")

class Mother:
  m = 70000
  def mother_method(self):
    print("mother class method")

  def joint_property(self):
    print('mother share')


class Son(Father, Mother): # multiple inheritance
  s = 298349
  def son_method(self):
    print('son class method')
```

```python
son = Son()
```

image.png

```python
son.joint_property()
```

```python
class Son(Mother, Father): # multiple inheritance
  s = 298349
  def son_method(self):
    print('son class method')
```

```python
son = Son()
```

```python
son.joint_property()
```
