```python
# global vs local variable -- based on scope the variables changes.
# global variables
a = 10
b = "Raj"
c = True

def test():
  print(a, b, c)

test()
```

```python
# local variables
def test1():
  # locals - the scope is inside of test1() function, can't be accessed outside of it.
  x = 100
  y = 200
  z = 300
  print(x, y, z)

test1()
```

```python
# how global variables can be accessed inside of a class
# inside of a class i am accessing the global variables
class GlobalTest:
  def test_global(self):
    print(a, b, c)

g = GlobalTest()
```

```python
class GlobalTest:
  def test_global(self):
    print(a, b, c)
    # a new global variable inside of a class
    global d
    d = 200
g = GlobalTest()
g.test_global()
```

```python
def test3():
  # inside of a class local variables will have high priority
  a = 1000
  b = 2000
  c = 3000
  d = 4000
  print(a, b, c, d)
  # what if i need global variables here?
  # global variables can't be accessed inside of a class directly --> you have use a function "globals()"
  print(globals()["a"], globals()["b"], globals()["c"], globals()["d"]) # 10 Raj True 200

test3()
```

```python
# instance variable creation -- 2 ways 1. inside of a constructor 2. inside of a instance method
class Student:
  def __init__(self, name):
    self.name = name # i am creating a instance variable - name

  def setAge(self, age):
    self.age = age # instance variable creation inside of a method.


s1 = Student("Narmadha")
print(s1.name)
# print(s1.age)
s1.setAge(25)
print(s1.age)
```

```python
# local variables
name = "Ram"
class Test:
  def local_variables(self):
    name = "Krishna"
    print(name)

t = Test()
t.local_variables()
```

```python
# method overloading
class Employee:
  def sleep(self):
    print("sleeping")

  def sleep(self, name):
    print('i am not sleeping anymore')

e = Employee()
e.sleep()
```

```python
# 1st way
# method overloading
class Employee:

  def sleep(self, name):
    print('i am not sleeping anymore')

  def sleep(self):
    print("sleeping")

e = Employee()
e.sleep()
```

```python
# 2nd way -- pass the argument
# method overloading - change in no.of arguments
class Employee:
  def sleep(self):
    print("sleeping")

  def sleep(self, name):
    print('i am not sleeping anymore')

e = Employee()
e.sleep("Shiv")

# overloading not possible because python is dynamically typed.
```
