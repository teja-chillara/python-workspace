```python
class Person:
  def __init__(self, name, age):
    print("i am inside a person constructor")
    self.name = name
    self.age = age

  # method
  def test(self):
    print('I am inside of a Person class')
```

```python
# variables - instance, static, local

# instance variable --> as long as, you see "self" that is instance

class Student:

  # outside of a method, inside of class --- static variable
  college_name = "JNTU"

  def __init__(self, student_name, student_city):
    # instance variables
    self.student_name = student_name
    self.student_city = student_city
```

```python
# self is depenendent on object creation
# 2. if you want to access instance variables you have to create object -- instance, object

# static is independent of object
Student.college_name
```

```python
# access -instance variables you have to create object of it
s1 = Student('Rishitha', "Bengaluru")
```

```python
s1.student_name
```

```python
s1.student_city
```

```python
print(s1.college_name) # static variables can be accessed through both ClassName or object
```
