```python
l1 = ['key1', 'key2', 'key3', 'key4']
v1 = 10

# {'key1': 10, 'key2':10, 'key3': 10, 'key4': 10}
result = {}
for key in l1:
  result[key] = v1

print(result)
```

```python
d11 = {}
d11.fromkeys(l1, v1)
```

```python
l1 = ['key1', 'key2', 'key3', 'key4']
v1 = (10, 20, 30, 40)

d12 = {}
d12.fromkeys(l1, v1)
```

```python
for key, value in zip(l1, v1):
  print(key, value)
```

```python
d13 = {}
for key, value in zip(l1, v1):
  d13[key] = value

print(d13)
```

```python
l2 = [1, 2, 3]
v2 = (10, 20)

for key, value in zip(l2, v2, strict=True):
  print(key, value)
```
