```python
class Test:
  # if self is involved i.e a instance method - these are depenedent on creation.
  def m1(self):
    print('i am m1() method')
```

```python
t = Test()
t.m1()
```

```python
class Test:
  def __init__(self):
    self.name = "Krishna"
    self.friend = "balaram"

  def m1(self):
    print('m1() method')
    print(self.name, self.friend)

t1 = Test()
t1.m1()

# you should go for instance methods, when it is dependent on object creation
# the only way you can access the instance methods, by creating an object
```

```python
Test.m1()
```

```python
# if self is involved you creating instance method
class Test:
  def m2(self):
    print("m2() method")
```

```python
class Test:
  @classmethod
  def m2(cls):
    print("m2() class method")
# you don't have to create object to access the class methods
```

```python
Test.m2()
```

```python
# in your class method, you can access all static variables.
class Test:
  # static variable
  a = 10
  b = 20

  # instance method
  # if inside of a instance method, you want to access the static variables use class name
  def m1(self):
    print('m1() instance method')
    print(Test.a, Test.b)

  @classmethod
  def m3(cls):
    print('class method')
    print(cls.a, cls.b)
    print(Test.a, Test.b)


Test.m3() # class method call.
print("___"*20)
t3 = Test() # object creation
t3.m1() # instance method with the help of object
```

```python
class StaticTest:
  @staticmethod
  def m7():
    print("m7() method")

StaticTest.m7()
```
