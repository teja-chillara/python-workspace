```python
# inheritance: Accessing the properties super/parent class to sub/child class is called inheritance.

# types of inheritance:
# 1. single level - one super and one sub class (father - son)
# 2. multi level - grand father -father - son - child
# 3. hierarchial inheritance - Tree
# 4. multiple inheritance - a single inheriting mutliple class (both father, mother)

# rules
# 1. a child class can access all the properties of a parent class (child can get fathers properties)
# 2. a super class can't access the sub class properties. (father can't access child properties)
```

# Single level inheritance

image.png

```python
class A: # A - class name
  a = 10 # static variable - a
  def __init__(self, x): # declaring a constructor with a parameterized
    self.x = x

  def a_test(self): # instance method
    print('a class test method')

  @classmethod
  def a_class(cls): # class method
    print('i am class method of a')

  @staticmethod
  def a_static(): # static method
    print('i am static method of A')
```

```python
class B(A): # A - superclass/parent; B - subclass/childclass
  pass
```

```python
a = A(200)
```

```python
a.a
```

```python
a.x
```

```python
a.a_test()
```

```python
# on your object reference you can call the static and class methods
a.a_class()
```

```python
a.a_static()
```

```python
A.a_class()
```

```python
A.a_static()
```

```python
B.a_static()
```

```python
B.a_class()
```

```python
b = B()
```

```python
b = B("B class constructor")
```

```python
b.a_test()
```

# Mutlti level inheritance

image.png

```python
class A: # can't access properties of B, C classes
  a = 10

class B(A): # B class can access A class, but not C class
  b = 20

class C(B): # C class can access A, B classes both.
  c = 30
```

```python
A.a
```

```python
A.b # here, A class can access "A" class properties only
```

```python
B.a
```

```python
B.b
```

```python
B.c
```

```python
C.a
```

```python
C.b
```

```python
C.c
```

# Hierachial inheritance
image.png

```python
class G: # supermost class
  g = "g"

class B(G):
  b = "b"

class E(G):
  e = "E"

class A(B): # from class A, you can access B, G class properties.
  a = "A"

class D(E): # i am not directly accessing class G, you are inheriting E class from E class you are accessing G class
  d = "D"
```

```python
A.a
```

```python
A.b
```

```python
A.g
```

# mutliple inheritance
image.png

```python
class A:
  a = 10
  x = 100

class B:
  b = 20
  x = 200

class C(A, B): # multiple inhertance (getting the properties of both father's and mother's)
  c = 30
```

```python
C.a
```

```python
C.b
```

```python
C.c
```
