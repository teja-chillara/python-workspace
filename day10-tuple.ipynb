```python
nums: list[str] = ['1', '2', '5', '7']

# output = [1, 2, 5, 7]

result = []
for num in nums:
  result.append(int(num))

print(nums)
print(result)
```

```python
[int(num) for num in nums] # list comprehension
```

```python
# syntax:
# [i for i in nums if condition]

nums = [1, 2, 3, 4, 5, 6, 7, 8, 9]
even_nums = []
for num in nums:
  if num % 2 == 0:
    even_nums.append(num)

print(even_nums)
```

```python
[num for num in nums if num % 2 == 0]
```

```python
[num for num in nums if num % 2 != 0]
```
