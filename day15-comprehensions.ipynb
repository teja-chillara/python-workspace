```python
nums = [1, 2, 3, 4, 5]
```

```python
even_nums = []
for num in nums:
  if num % 2 == 0:
    even_nums.append(num)

print(even_nums)
```

```python
# list comprehensions
[num for num in nums if num % 2 == 0]
```

```python
[num**2 for num in nums] # [1, 4, 9, 16, 25]
```

```python
# list comprehension
[num for num in nums if num % 2 != 0] # odd
```

```python
# tuple comprehension
tuple((num for num in nums if num % 2 != 0)) # tuple comprehension isn't possible, even though you use () you will generator object
# again you convert that generator object to tuple (using type conversion)
```

```python
# set comprehension
{num for num in nums if num % 2 != 0}
```

```python
result = {num**2 for num in nums}
result
```

```python
type(result)
```

```python
# dict comprehension   # syntax: {key:value for loop if condition}
{num:num**2 for num in nums}
```

```python
{"num":num**2 for num in nums} # {"num": 25}
```
