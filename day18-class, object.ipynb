```python
class Student:
  # construction
  def __init__(self, name, age, city):
    print("i am inside of a constructor")
    # self - current object
    self.name = name
    self.age = age
    self.city = city
```

```python
student1 = Student()
```

```python
student1 = Student("Ram", 25, "Chennai")
```

```python
student1.name
```

```python
student1.age
```

```python
student1.city
```
