```python
persons = ['student', 'engineer', 'police', 'theif', 'MLA', 'graduate', 'farmer']
```

```python
# jumping/transfer statements (continue, break)
for person in persons:
  print(person)
```

```python
for person in persons:
  if person == 'theif':
    print('found the theif')
  else:
    print('continue the search')
```

```python
for person in persons:
  if person == 'theif':
    print('found the theif')
    break
  else:
    print('continue the search')
```

```python
for person in persons:
  if person == 'MLA':
    continue
  else:
    print('collect toll from', person)
```
