# conditional statments

to check/validation

syntax:


```
if condition:
  print('if condition')
else:
  print('else condition')
```

Rule:
1. with out if block, else block can't be existed. (if condition is mandatory)
2. if "if" condition get's true, only if condition will be executed. else will not.
3. if condition fails, then only else block will get execute.
4. only if condition will not syntax error (no else block)


````python
condition = True

if condition: # if condition is true only if block will get execute, else block will be ignored.
  print('first line of if')
  a = 10
else:
  print('this is else block')
````

````python
if condition:
  print('only if condition')
````




````python
condition1 = True
condition2 = True
condition3 = True

# if all conditions are true, then execute it
if condition1 and  condition2 and condition3: # 1 and 0 --> 0 and 1 --> 0 # 1 and 1 --> 1 and 1 --> 1
  print('all condtion are true')
else:
  print('all conditions are failed')
````

````python
condition2 = False
if condition1 or condition2: # 1 or 0 --> 1
  print('if block')
else:
  print('else block')
````
